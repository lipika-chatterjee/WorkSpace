#Data Cleaning

```python
from IPython.core.interactiveshell import dis
import pandas as pd
from collections import Counter
# 1. Convert the "stringified" lists into actual Python lists
import ast
#games_data['Genres'] = games_data['Genres'].apply(ast.literal_eval)
# 2. Now explode them into individual items and convert to a list
#all_genres_list = games_data['Genres'].explode().tolist()
# View the result
#genres_list=(list(set(all_genres_list)))
#counts = Counter(games_data['Genres'].explode().tolist())
#counts
#Adventure 1014
#RPG 523
#Platform 329
#Shooter 353
#Indie 284
games_data['Adventure']=0
games_data['RPG']=0
games_data['Platform']=0
games_data['Shooter']=0
games_data['Indie']=0
for i in range(len(games_data['Genres'] )):
  if 'Adventure' in (games_data.loc[i,'Genres']):
    games_data.loc[i,'Adventure']=1
  if 'RPG' in (games_data.loc[i,'Genres'] ):
    games_data.loc[i,'RPG']=1
  if 'Platform' in (games_data.loc[i,'Genres'] ):
    games_data.loc[i,'Platform']=1
  if 'Shooter' in (games_data.loc[i,'Genres'] ):
    games_data.loc[i,'Shooter']=1
  if 'Indie' in (games_data.loc[i,'Genres'] ):
    games_data.loc[i,'Indie']=1
games_data[['Adventure','RPG','Platform','Shooter','Indie']]
```

### remove duplicate and handle null values in ratings,plays and sales.

```python
for i in range(len(games_data['Plays'])):
  if 'K' in games_data.loc[i]['Plays']:
    games_data.loc[i,'Plays']=float(games_data.loc[i,'Plays'][0:-1])*1000
```

```python
for i in range(len(games_data['Times Listed'])):
  if 'K' in (games_data.loc[i,'Times Listed']):
    games_data.loc[i,'Times Listed']=(float(games_data.loc[i,'Times Listed'][0:-1])*1000)
```

```python
for i in range(len(games_data['Number of Reviews'])):
  if 'K' in (games_data.loc[i,'Number of Reviews']):
    games_data.loc[i,'Number of Reviews']=(float(games_data.loc[i,'Number of Reviews'][0:-1])*1000)
```

```python
for i in range(len(games_data['Playing'])):
  if 'K' in (games_data.loc[i,'Playing']):
    games_data.loc[i,'Playing']=(float(games_data.loc[i,'Playing'][0:-1])*1000)
```

```python
for i in range(len(games_data['Backlogs'])):
  if 'K' in (games_data.loc[i,'Backlogs']):
    games_data.loc[i,'Backlogs']=(float(games_data.loc[i,'Backlogs'][0:-1])*1000)
```

```python
for i in range(len(games_data['Wishlist'])):
  if 'K' in (games_data.loc[i,'Wishlist']):
    games_data.loc[i,'Wishlist']=(float(games_data.loc[i,'Wishlist'][0:-1])*1000)
```
